**Analysing Gender**

Lets analyze the ratio of the purchase made by each gender.

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]},
                    index = ['Male', 'Female'])

data.plot.pie(y= 'Ratio' , figsize = (6,6), autopct = "%.1f")
```

We can also do this by using groupby function.

```python
df.groupby('Gender').size().plot(kind='pie', autopct='%.1f', title= 'Gender Ratio',figsize= (6,6))
```

Let's plot it on a bar graph.

```python
df.groupby('Gender').size().plot(kind='bar', figsize= (6,6) )
```

```python
df.groupby('Gender').size()
```

It is clear that men purchased more things but lets check the pize of the purchase they both made.
